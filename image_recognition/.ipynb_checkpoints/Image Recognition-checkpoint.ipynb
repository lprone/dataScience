```python
%%time
import tensorflow as tf

from imageai.Detection import ObjectDetection

import os
from IPython.display import Image,display

import fnmatch
```

```python
execution_path = os.getcwd()
```

```python
detector = ObjectDetection()
detector.setModelTypeAsRetinaNet()
```

```python
detector.setModelPath( os.path.join(execution_path , "resnet50_coco_best_v2.0.3.h5"))
```

```python
%%time
detector.loadModel()
```

```python
%%time
for image_name in fnmatch.filter(os.listdir('.'), '*.jpg'):
    print(image_name)
    display(Image(image_name))
    new_image_name = 'new' + image_name
    detections = detector.detectObjectsFromImage(input_image=os.path.join(execution_path , image_name), output_image_path=os.path.join(execution_path , new_image_name))
    display(Image(new_image_name))
    for eachObject in detections:
        print(eachObject["name"] + " : " + eachObject["percentage_probability"] )
    os.remove(new_image_name)
    print('#################################################################################################')  
```
